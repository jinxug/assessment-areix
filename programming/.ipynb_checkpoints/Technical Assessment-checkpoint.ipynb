# Task 1 - Dynamic Programming

```python
#set up sorted satisfaction and time lists
def task1(satisfaction=[]):
    print("Input satisfaction = "+str(satisfaction))
    satisfaction.sort()
    length = len(satisfaction)
    time = range(1,length+1)
    result = 0
    


    #find the maximum product
    if satisfaction[-1] <= 0:
        print("Explanation: People don't like the woodworks. No woodwork is made")
    else:    
        for i in range(length):
            multi_list = [satisfaction[j]*time[j-i] for j in range(i,length)]
            #print(multi_list)
            total = 0
            for k in multi_list:
                total = total+k
            if total>result:
                result = total

    print("Output: "+str(result)+"\n")
    
```

```python
task1([-1,-8,0,5,-9])
task1([4,3,2])
task1([-1,-4,-5])
task1([-2,5,-1,0,3,-3])
```

# Task 2 - Stack, 2 pointers

```python
def water_i(i,a =[]):
    left = a[0:i+1]
    right = a[i:len(a)]
    return min(max(left),max(right))

def task2(a=[]):
    print("Input: "+ str(a))
    length = len(a)
    water = []

    for i in range(length):
        water_acc = water_i(i,a)-a[i]
        water.append(water_acc)

    total = 0
    for k in water:
        total = total+k
    print("Output: "+ str(total))
```

```python
task2([0,1,0,2,1,0,1,3,2,1,2,1])
```

```python
MFName = []
for op in soup.find_all('select', {'class' : 'select', 'id' : 'NavDownMFName'})[0]:
    str_op = str(op)[15:].split('">')
    if len(str_op) > 1 and str_op[0] != '':
        MF_Value = str_op[0] 
        MF_Name = str_op[1].split('<')[0]
        MFName.append(MF_Value +' ' + MF_Name)
MFName
```

```python
response = requests.get(url)
data = {}
sta = response.text.split(Name)
iterate = 1
for i in range(1,len(sta)):
    if i != (len(sta)-1):
        #part one
        current = sta[i].split('\n')
        current.pop()
        current.pop()
        current.pop()
        current.pop()
        for j in current:
            if j != '\r' and j!='':
                datum = {}
                j_data = j.split(';')
                datum['id'] = j_data[0]
                datum['plan'] = j_data[1]
                datum['word'] = j_data[2]
                datum['price'] = j_data[4]
                datum['date'] = j_data[7]
                data[iterate] = datum
                iterate = iterate+1
                
    else:
        #last part
        current = sta[i].split('\n')
        for j in current:
            if j != '\r' and j!= '':
                datum = {}
                j_data = j.split(';')
                datum['id'] = j_data[0]
                datum['plan'] = j_data[1]
                datum['word'] = j_data[2]
                datum['price'] = j_data[4]
                datum['date'] = j_data[7]
                data[iterate] = datum
                iterate = iterate+1
```

```python
#output data into text file as Json format
data_json = json.dumps(data)
with open('./data.txt', 'w')as json_file:
    json.dump(data_json,json_file)
```

## 1. get train and test data

```python
%matplotlib inline
import IPython.core.display         
# setup output image format (Chrome works best)
IPython.core.display.set_matplotlib_formats("svg")
import matplotlib.pyplot as plt
import matplotlib
from numpy import *
from sklearn import *
import glob
import os
import csv
import string
random.seed(100)
```

```python
train_filename = './train.csv'

# read the data
allfeatnames = []
textdata      = []
with open(train_filename, 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        if len(allfeatnames)==0:
            allfeatnames = row
        else:
            alert = 0
            for x in row:
                if x == '':
                    alert = 1
            if alert == 0:
                textdata.append(row)

#print(allfeatnames)
#print(textdata)

# put the data into a np array
dataX = empty((len(textdata), len(allfeatnames)-1))  #build a matrix
dataY = empty(len(textdata)) #build an array
for i,row in enumerate(textdata):
    #print(row)
    # extract features (remove the first 2 features and the last feature)
    dataX[i,:] = array([float(x) for x in row[1:]])    #neglect the first 2 and the last one
    # extract target (last entry)
    dataY[i] = row[0]
    
# extract feature names
featnames = [x.strip() for x in allfeatnames[1:]]
print(dataX.shape)
print(dataY.shape)
trainX = dataX
trainY = dataY
```

```python
test_filename = './test1.csv'
allfeatnames = []
textdata      = []
with open(test_filename, 'r') as f2:
    reader = csv.reader(f2)
    for row in reader:
        #print(row)
        if len(allfeatnames)==0:
            allfeatnames = row
        else:
            alert = 0
            for x in row:
                if x == '':
                    alert = 1
            if alert == 0:
                textdata.append(row)

#print(allfeatnames)
#print(textdata)

# put the data into a np array
dataX = empty((len(textdata), len(allfeatnames)))  #build a matrix
dataY = empty(len(textdata)) #build an array
for i,row in enumerate(textdata):
    #print(row)
    # extract features (remove the first 2 features and the last feature)
    dataX[i,:] = array([float(x) for x in row])    #neglect the first 2 and the last one
    # extract target (last entry)
    dataY[i] = row[0]
    
# extract feature names
featnames = [x.strip() for x in allfeatnames]
print(dataX.shape)
print(dataY.shape)
testX = dataX
testY = dataY
print(testX)
print(testY)
```

## 2. normalization

```python
# normalize feature values
# this makes comparing weights more meaningful
scaler = preprocessing.StandardScaler()  
trainXn = scaler.fit_transform(trainX)  
testXn  = scaler.transform(testX) 
#print("OLS: train error =", trainAE)
```

## 3. training with logistic regression  

```python
from sklearn.linear_model import LogisticRegressionCV
clf = LogisticRegressionCV(Cs = logspace(-4,4,20),cv = 5)
clf.fit(trainXn, trainY)
trainAE = metrics.mean_absolute_error(trainY, clf.predict(trainXn))
print("Logistic Regression: absolute trainning error =", trainAE)
print('Predicted Label: '+str([x for x in clf.predict(testXn)]))
```

## 4. training with Naive Bayes

```python
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import BernoulliNB

clf = BernoulliNB(alpha = 0.1)
bmodel=clf.fit(trainXn,trainY)
trainAE = metrics.mean_absolute_error(trainY, clf.predict(trainXn))
print("Naive Bayes: absolute trainning error =", trainAE)
print('Predicted Label: '+str([x for x in clf.predict(testXn)]))
```

## 5. training with KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

knn = KNeighborsClassifier()
knn.fit(trainXn,trainY)
trainAE = metrics.mean_absolute_error(trainY, knn.predict(trainXn))
print("KNN: absolute trainning error =", trainAE)
print('Predicted Label: '+str([x for x in knn.predict(testXn)]))
```

## 6.training with linear regression


```python
from sklearn.linear_model import LinearRegression
ols = LinearRegression()
ols.fit(trainXn,trainY)
trainAE = metrics.mean_absolute_error(trainY, ols.predict(trainXn))
print("OLS Linear Regression: absolute trainning error =", trainAE)
print('Predicted Label: '+str([round(x) for x in ols.predict(testXn)]))
```

## 7. training with Ridge Regression

```python
from sklearn.linear_model import RidgeCV
rr = RidgeCV(alphas = logspace(-4,4,20))
rr.fit(trainXn,trainY)
trainAE = metrics.mean_absolute_error(trainY, rr.predict(trainXn))
print("Ridge Regression: absolute trainning error =", trainAE)
print('Predicted Label: '+str([round(x) for x in rr.predict(testXn)]))
```
